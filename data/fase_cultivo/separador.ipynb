```python
import pandas as pd
```

```python
df = pd.read_csv('Maduracion.csv')
df.head()
```

```python
columnas_a_dividir = [
    'Temp. Recomendada (°C)',
    'Lluvia Óptima (mm)',
    'Humedad Ambiental (%)',
    'Humedad Suelo (%)'
]
```

```python
for col in columnas_a_dividir:

    split_data = df[col].str.split('-', expand=True)
    
    df[f'{col}_valor_mínimo'] = split_data[0]
    df[f'{col}_valor_máximo'] = split_data[1]
```

```python
print("Primeras filas con las nuevas columnas:")
display(df.head())
```

```python
cols_a_borrar = [
    'Fase',
    'Fuente URL',
    'Temp. Recomendada (°C)', 
    'Lluvia Óptima (mm)', 
    'Humedad Ambiental (%)', 
    'Humedad Suelo (%)'
]

df.drop(columns=cols_a_borrar, inplace=True)
df.head()
```

```python
# 1. Identificamos automáticamente las columnas que tienen "_valor_mínimo"
columnas_minimas = [c for c in df.columns if c.endswith('_valor_mínimo')]

# 2. Recorremos cada una para calcular el promedio
for col_min in columnas_minimas:
    # Obtenemos el nombre base (quitando '_valor_mínimo')
    nombre_base = col_min.replace('_valor_mínimo', '')
    
    # Construimos el nombre de la columna máxima correspondiente
    col_max = nombre_base + '_valor_máximo'
    
    # Verificamos si existe la pareja (columna máxima)
    if col_max in df.columns:
        # Convertimos a números (float) para poder hacer matemáticas
        # errors='coerce' convierte en NaN (vacío) si hay algún texto que no sea número
        val_min = pd.to_numeric(df[col_min], errors='coerce')
        val_max = pd.to_numeric(df[col_max], errors='coerce')
        
        # Calculamos el promedio y creamos la nueva columna
        nombre_promedio = nombre_base + '_valor_promedio'
        df[nombre_promedio] = (val_min + val_max) / 2

# 3. Vemos el resultado final con las nuevas columnas de promedio
print("Nuevas columnas de promedio generadas:")
cols_promedio = [c for c in df.columns if '_valor_promedio' in c]
display(df[cols_promedio].head())
```
